```python
# Load Data from CSV file

import numpy as np

a = [('name','S32'),('Age','i4'),('Gender','S32'),('Department','S32'),('Year_of_study','i4'),
         ('Recent_CGPA','f4'),('CGPA_status','S32'),('Hours_spent','i4'),('Purpose','S32'),('Devices_Using','i4'),
     ('Hours_an_year','i4')]

data = np.genfromtxt('ElectronicGadgets.csv', delimiter = ',', dtype = a)

# Modified first name by creating a copy of the data

x = data.copy()
data[0][0] = b"Aruna"

copyDataTitle = 'Copy of the Data'
title = copyDataTitle.upper().center(127, ':')
print(title)
print(x, end ='\n\n')

modifiedDataTitle = 'Modified Data'
title = modifiedDataTitle.upper().center(127, ':')
print(title)
print((data),end = '\n\n')
print("Total Number of records in the actual data are ", len(data),end ='\n\n')
```

```python
# Add coloumn using formula

data['Hours_an_year'] = data['Hours_spent'] * 365
#np.savetxt('ElectronicGadgets.csv', data, delimiter = ',', fmt = '%s')
print(data)
```

```python
# User entry

while True:
    userInput = input('Want to enter data? (Yes or No):')
    if userInput.lower() == 'yes':
        name = input("Enter name: ")
        name = name.title()
        Age = int(input("Enter Age: ")) 
        Gender = input("Enter Gender: ")
        Gender= Gender.title()
        Dept = input("Enter Department: ")
        Dept = Dept.title()
        Year = int(input("Enter Year: "))
        Recent_CGPA = float(input("Enter CGPA: ")) 
        status = input("Enter status: ")
        status = status.title()
        hours = int(input("Enter hours spent: "))
        purpose = input("Enter purpose: ")
        purpose = purpose.title()
        number = int(input("Enter number of devices using? ")) 
        hours_an_year = hours*365
        n = np.array([(name, Age, Gender, Dept, Year, Recent_CGPA, status, hours, purpose, number,hours_an_year)], 
                       dtype = a) 
        data = np.append(data, n)
        np.savetxt('ElectronicGadgets.csv', data, delimiter = ',', fmt = '%s')
        print(data) 
    elif userInput.lower() == 'no':
        break
    else:
        print('#Invalid entry.\nTry again.')
```

```python
#Finding Unique values

headerArray = ['name', 'Age', 'Gender', 'Department', 'Year_of_study','Recent_CGPA',
               'CGPA_status', 'Hours_spent', 'Purpose', 'Devices_Using', 'Hours_an_year']

for i in range(0, len(headerArray)):
    headerArrayTitle = headerArray[i].replace("_", " ").upper()
    title = headerArrayTitle.center(127, ':')
    print(title)
    print(np.unique(data[headerArray[i]]), end = '\n\n')
```

```python
# Replacing the missing values with NA if it a string and 0 if it is -1

i = 0
missing_data = data
while i < len(data):
    j = 0
    while j < len(data[i]):
        if data[i][j] == b'' or data[i][j] == b' ':
            missing_data[i][j] = b'NA'
        else:
            if data[i][j] == -1 or data[i][j] == -365:
                missing_data[i][j] = 0
        j = j+1
    i = i+1

headerArrayTitle = 'Filled missing values'
title = headerArrayTitle.upper().center(127, ':')
print(title)
# np.savetxt('ElectronicGadgets.csv', missing_data, delimiter=',', fmt='%s')
print(missing_data)
```

```python
# Repalcing Department names with proper names

i = 0
while i < len(missing_data):
    j = 0
    while j < len(missing_data[i]):
        if missing_data[i][j] == b'betch': 
            missing_data[i][j] = b'Btech'
        else:
            if missing_data[i][j] == b'Mbbs': 
                missing_data[i][j] = b'MBBS'
            else:
                if missing_data[i][j] == b'Bds':
                    missing_data[i][j] == b'BDS'
                else:
                    if missing_data[i][j] == b'pharmacy': 
                        missing_data[i][j] = b'Pharmacy'
        j = j+1
    i = i+1
dataArrayTitle = 'Replaced improper Department Names'
title = dataArrayTitle.upper().center(127, ':')
print(title) 
print(missing_data)
```

```python
# Delete the rows which has NA or 0 using For loop

clone_data = np.copy(missing_data)
for i in range(len(missing_data)-1, -1, -1):
    if b'NA' in missing_data[i] or 0 in missing_data[i]:
        clone_data = np.delete(clone_data, i, axis = 0)
            
np.savetxt('ElectronicGadgets.csv', clone_data, delimiter = ',', fmt = '%s')
cloneDataArrayTitle = 'Clean Data'
title = cloneDataArrayTitle.upper().center(127, ':')
print(title) 
print(clone_data)
print('\nNumber of records in clean data', len(clone_data))
```

```python
#Delete the rows which has NA or 0 using while loop

clone_data1 = np.copy(missing_data)
i = len(missing_data) - 1

while i >= 0:
    if b'NA' in missing_data[i] or 0 in missing_data[i]:
        clone_data1 = np.delete(clone_data1, i, axis=0)
    i = i-1

cloneDataArrayTitle = 'Clean Data'
title = cloneDataArrayTitle.upper().center(127, ':')
print(title) 
print(clone_data1)
print('\nNumber of records in clean data', len(clone_data1))
```

```python
# Sorting

clean_data = np.copy(clone_data1)
clean_data = np.sort(clone_data1, order = ['Gender'])
gender = 'Sorting based on Gender'
title = gender.upper().center(127, ':')
print(title) 
print(clean_data, end = '\n\n')

# Masking and Extracting data For Females

maskfemales = (clean_data['Gender']==b'Female')
females = clean_data[maskfemales]
gender = 'Female Data'
title = gender.upper().center(127, ':')
print(title)
print(females)
print("\nTotal Number of Females are ", len(females), end = '\n\n')

# Masking and Extracting data For Males

maskmales = (clean_data['Gender'] == b'Male')
males = clean_data[maskmales]
gender = 'Male Data'
title = gender.upper().center(127, ':')
print(title)
print(males)
print("\nTotal Number of Males are ", len(males))
```

```python
# Get the entire details of the student from the data when given their name

userinput = eval(input("Enter name: "))
userinput = userinput.title()
Index = np.where(userinput == data['name'])
print(Index)
Details = data[Index]

information = ("Name of the the Student is: ","Age of the student is: ","Gender: ","Department: ","Year of Study: ",
               "Recent CGPA for 4: ","CGPA_status: ","Hours spent on electronic gadgets in a day: ",
               "Purpose of using gadgets: ","Number of Devices using: ","Hours in an year that he/she spends on gadgets: ")
i = 0
while i == 0 :
    j = 0
    while j < len(Details[i]) :
        print(information[j], Details[i][j])
        j = j+1
    i = i+1
```
